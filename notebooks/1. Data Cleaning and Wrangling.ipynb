## Transforming Columns

```python
df_movies[df_movies.duplicated()].sort_values(by="title") # checking duplicates to be dropped
```

```python
df_movies.drop_duplicates(inplace=True) 
```

```python
import ast

ast.literal_eval(df_movies["belongs_to_collection"][0]) # convert string elements in this column to dictionaries
```

```python
# converting all elements in "belongs_to_collection" column to dictionaries

df_movies["collection"] = [x if pd.isnull(x) else ast.literal_eval(x) for x in df_movies["belongs_to_collection"]]
```

```python
df_movies["collection"][0].get("name") # get the value using the key
```

```python
# new column that will replace "belong_to_collection" only need the name of the collection

df_movies["collection"] = [x.get("name") if type(x) == dict else np.nan for x in df_movies["collection"]] 
```

The following columns are lists of dictionaries, so they will need slightly different treatment.

```python
df_movies["genres_list"] = [x if pd.isnull(x) else ast.literal_eval(x) for x in df_movies["genres"]]
```

```python
df_movies["genres_list"][0] 
```

```python
df_movies["genres_list"][0][0]["name"]
```

```python
# for every element in these columns, we do a comprehension list inside of a comprehension list
# every cell will have a list with genres

df_movies["genres_list"] = [[genre["name"] for genre in element] for element in df_movies["genres_list"]]
```

Repeating this process for other columns with similar structure.

```python
# converting strings to lists

df_movies["spoken_languages_list"] = [x if pd.isnull(x) else ast.literal_eval(x) for x in df_movies["spoken_languages"]]
df_movies["production_companies_list"] = [x if pd.isnull(x) else ast.literal_eval(x) for x in df_movies["production_companies"]]
df_movies["production_countries_list"] = [x if pd.isnull(x) else ast.literal_eval(x) for x in df_movies["production_countries"]]
```

```python
# this comprehension list check first if the element is a list (list of dictionaries), if it's not a list, it will be converted to null value

df_movies["spoken_languages_list"] = [[language["iso_639_1"] for language in element] if type(element) == list else np.nan for element in df_movies["spoken_languages_list"]]
```

```python
df_movies["production_companies_list"] = [[company["name"] for company in element] if type(element) == list else np.nan for element in df_movies["production_companies_list"]]
```

```python
df_movies["production_countries_list"] = [[country["iso_3166_1"] for country in element] if type(element) == list else np.nan for element in df_movies["production_countries_list"]]
```

Now we drop the columns we won't goint to use.

```python
discard_columns = ["video","imdb_id","adult","original_title","poster_path", # these columns won't be used to feed the machine learning model
                   "homepage","belongs_to_collection", "tagline", # tagline has a lot of empty values
                   "genres", "production_companies",  # these columns were replaced for new ones, so they won't be needed
                   "production_countries",
                    "spoken_languages" ]
```

```python
df_movies.drop(columns=discard_columns, inplace=True)
```

## Handling Missing Values and Adding Columns

```python
df_movies.isna().mean()
```

```python
# there are non numerical values in budget column so it can not be converted to float
# also the release dates are in the id column and the id's are in the release_date column
# popularity and genres_list are also missplaced

df_movies[df_movies['budget'].str.contains(r'\D')]  # regex to identify numerical rows in budget
```

```python
def convert_to_float(value): 
    try:
        return float(value) 
    except (ValueError, TypeError): # if there is an error, it will replace the strange values for null values
        return np.nan
```

```python
df_movies["budget"] = [convert_to_float(x) for x in df_movies["budget"]] # now every value here is float
```

```python
df_movies["budget"].fillna(0, inplace=True) # there is no missing values anymore on budget and revenue
df_movies["revenue"].fillna(0, inplace=True)
```

```python
df_movies['release_date'].sort_values() # there are values that we can not convert in date type 
```

```python
def convert_to_datetime(value): 
    try:
        return pd.to_datetime(value, format="%Y-%m-%d") 
    except (ValueError, TypeError):  
        return np.nan # the data that can not be converted to datetime will be converted to null
```

```python
df_movies['release_date'] = [convert_to_datetime(x) for x in df_movies['release_date']]
```

```python
# drop rows only if there are null values in "release_date" column
# dropping this rows, also delete the table above with values missplaced, so we don't need to treat them anymore
# there were only three values of dozens of thousands, so it won't affect further analysis

df_movies.dropna(subset=['release_date'], inplace=True) 
```

```python
df_movies["release_year"] = df_movies["release_date"].dt.year
df_movies["release_year"]
```

```python
df_movies["return"] = np.where(df_movies["budget"] != 0, df_movies["revenue"] / df_movies["budget"], 0)
```

# Creating Extra Datasets

To complete this project, movies, cast and crew are more than enough, but building the API functions will be a lot easier with two extra tables. The First one will be financial data per actor.

```python
df_movies = pd.read_csv("../processed_data/movies.csv")
```

```python
df_cast = pd.read_csv("../processed_data/cast.csv")
df_crew = pd.read_csv("../processed_data/crew.csv")
```

```python
# merging the cast and movies dataset to get the financial information of every actor

join_actor = df_cast.merge(df_movies[["id", "title", "budget", "revenue", "return"]], how="left", on="id")  
join_actor
```

```python
# grouping the names of directors to get the total budget and total revenue

budget_actor = join_actor.groupby("name")["budget"].sum().sort_values(ascending=False).reset_index() # needs reset_index() to convert it to pandas dataframe
revenue_actor = join_actor.groupby("name")["revenue"].sum().sort_values(ascending=False).reset_index()

# merge this two new df's so all the info is in one table only
actor_financial = budget_actor.merge(revenue_actor, how="left", on="name")
actor_financial
```

```python
# total return = total revenue / total budget, except if budget is zero

actor_financial["total_return"] = np.where(actor_financial["budget"] != 0, actor_financial["revenue"] / actor_financial["budget"], 0)
```

```python
# grouping the actor's name but we calculate the mean return using every movie they've participated

actor_avg_return = join_actor.groupby("name")["return"].mean().round(2).sort_values(ascending=False).reset_index()
actor_avg_return
```

```python
# merging the table with average return with the main table

actor_financial = actor_financial.merge(actor_avg_return, how="left", on="name")
```

```python
# grouping by name to get the number of films of every actor

movies_actor = join_actor.groupby("name")["title"].count().sort_values(ascending=False).reset_index()
movies_actor
```

```python
# merging the table with number or movies with the main table

actor_financial = actor_financial.merge(movies_actor, how="left", on="name")
```

```python
# renaming columns

actor_financial = actor_financial.rename(columns={'budget': 'total_budget', 'revenue': 'total_revenue', 'return': 'average_return', 'title': 'films'})
actor_financial
```

Now getting financial data per director, doing similar transformations like the previous ones.

```python
director_df = df_crew[df_crew["job"] == "Director"] # filtering jobs
```

```python
join_director = director_df.merge(df_movies[["id", "title", "budget", "revenue", "return"]], how="left", on="id") 
budget_director = join_director.groupby("name")["budget"].sum().sort_values(ascending=False).reset_index()
revenue_director = join_director.groupby("name")["revenue"].sum().sort_values(ascending=False).reset_index()
director_financial = budget_director.merge(revenue_director, how="left", on="name")
```

```python
director_financial["total_return"] = np.where(director_financial["budget"] != 0, director_financial["revenue"] / director_financial["budget"], 0)
director_financial = director_financial.rename(columns={'budget': 'total_budget', 'revenue': 'total_revenue'})
director_financial
```
